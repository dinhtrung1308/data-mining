```python
a = pd.read_excel('./a.xlsx', header=None)
a.head()
abc = []
for i in range(0, 251):
    abc.append([str(a.values[i,j]) for j in range(0, 3)])
print(abc)

```

```python
print(a.head())
```

```python
association_rules = apriori(abc , min_confidence=0.8, min_length=2)
association_results = list(association_rules)
```

```python
print(len(association_results))
```

```python
print(association_results)
```

```python
vnstock=pd.read_excel('./vnstock.xlsx',header=None)
vnstock.head()
```
